```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NYPD_Complaint_Data_Historic2.csv")
```

```python
df.head().T
```

```python
df = df.drop(columns=['CMPLNT_NUM', 'ADDR_PCT_CD', 'BORO_NM', 'CMPLNT_FR_DT', 'CMPLNT_FR_TM', 'CMPLNT_TO_DT', 'CMPLNT_TO_TM', \
                      'CRM_ATPT_CPTD_CD', 'HADEVELOPT', 'HOUSING_PSA', 'JURISDICTION_CODE', 'JURIS_DESC', 'PARKS_NM', \
                      'PATROL_BORO', 'RPT_DT', 'STATION_NAME', 'TRANSIT_DISTRICT', 'X_COORD_CD', \
                      'Y_COORD_CD', 'Latitude', 'Longitude', 'Lat_Lon']) # 'PD_CD', 'PD_DESC',
df = df.astype({"KY_CD": int, "OFNS_DESC": 'category', "LAW_CAT_CD": 'category', "LOC_OF_OCCUR_DESC": 'category', \
               "PREM_TYP_DESC": 'category', "SUSP_AGE_GROUP": 'category', "SUSP_RACE": 'category', "SUSP_SEX": 'category', \
               "VIC_AGE_GROUP": 'category', "VIC_RACE":'category', "VIC_SEX": 'category', "PD_CD": 'category', \
                "PD_DESC": 'category'})
```

```python
df = df.dropna()
df = df.drop(df[df.values == 'UNKNOWN'].index)
```

```python
df = df.drop(df[~df.SUSP_AGE_GROUP.isin(['<18', '25-44', '18-24', '45-64', '65+'])].index)
df = df.drop(df[~df.VIC_AGE_GROUP.isin(['<18', '18-24', '25-44', '45-64', '65+'])].index)
df = df.drop(df[df.VIC_SEX.isin(['D', 'E'])].index)
df = df.drop(df[df.SUSP_SEX.isin(['U'])].index)
```
